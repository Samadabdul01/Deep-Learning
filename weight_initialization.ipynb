```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons
import tensorflow
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

```python
X, y = make_moons(n_samples = 100, noise = 0.2, random_state= 2)
```

```python
X.shape
```

```python
plt.scatter(X[:,0],X[:, 1], c = y)
```

```python
# Plot histogram for each feature to check distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].hist(X[:, 0], bins=20, color='skyblue', edgecolor='black')
axes[0].set_title('Feature 1 Distribution')
axes[0].set_xlabel('Feature 1')
axes[0].set_ylabel('Frequency')

axes[1].hist(X[:, 1], bins=20, color='salmon', edgecolor='black')
axes[1].set_title('Feature 2 Distribution')
axes[1].set_xlabel('Feature 2')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
model = Sequential()
model.add(layers.Dense(10, input_shape=(2,), activation='relu', kernel_initializer='he_normal'))
model.add(layers.Dense(10,  activation='relu', kernel_initializer='he_normal'))
model.add(layers.Dense(10,  activation='relu', kernel_initializer='he_normal'))
model.add(layers.Dense(1,  activation='sigmoid'))
```

```python
model.compile(optimizer = 'adam', loss = 'binay_crossentropy', metrics=['accuracy'])
```
